# json text_splitting

```python
import json

import requests

# This is a large nested json object and will be loaded as a python dict
json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

splitter = RecursiveJsonSplitter(max_chunk_size=300)

json_chunks = splitter.split_json(json_data=json_data)

for chunk in json_chunks[:3]: # taking first 3 items
    print(chunk)
```

```python
# The splitter can also output documents
docs = splitter.create_documents(texts=[json_data])

for doc in docs[:3]:
    print(doc)
```

# Vector Store

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

doc = TextLoader("../data/DiseaseText.txt", encoding="utf-8")
docmnet = doc.load()
txt_splitter = CharacterTextSplitter(chunk_size = 500, chunk_overlap=30)
doc_splitted = txt_splitter.split_documents(docmnet)
```

```python
#embedding
embeddings = OllamaEmbeddings(model="nomic-embed-text")
db = FAISS.from_documents(doc_splitted, embeddings)
db
```

```python
query = "Who gets giardiasis disease?"

docs = db.similarity_search(query)
docs[0].page_content
```

# as Retriver

```python
retriver = db.as_retriever()
ret = retriver.invoke(query)
ret[0].page_content
```

# Similarity search score

```python
docs_similarity_score = db.similarity_search_with_score(query)
docs_similarity_score
```

```python
embedding_vector = embeddings.embed_query(query)
db.similarity_search_by_vector(embedding_vector)
```

# saveing and laoding

```python
db.save_local("faiss_index")
```

```python
new_df = FAISS.load_local("./faiss_index", embeddings, allow_dangerous_deserialization=True)
docs = new_df.similarity_search(query)
for d in docs:
    print(d.page_content[:300])
```

# chroma db

```python
from langchain_chroma import Chroma

from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
data = TextLoader("../data/DiseaseText.txt", encoding="utf-8")
txt = data.load()
```

```python
txt_splitter = CharacterTextSplitter(chunk_size = 500, chunk_overlap=30)
doc_splitted = txt_splitter.split_documents(txt)
```

```python
embedding = OllamaEmbeddings(model="nomic-embed-text")
vectordb =  Chroma.from_documents(documents=doc_splitted, embedding=embedding)
```

```python
query = "Who gets giardiasis disease?"

vectordb_similarty = vectordb.similarity_search(query)
vectordb_similarty[0].page_content
```

```python
#save

vectordb =  Chroma.from_documents(documents=doc_splitted, embedding=embedding, persist_directory="../data/chromaStoreage")
```

```python
#load
db_load = Chroma(persist_directory="../data/chromaStoreage", embedding_function=embedding)
db_load_doc = db_load.similarity_search(query=query)
db_load_doc[0].page_content
```

```python
# as retriver
retriver = vectordb.as_retriever()
ret = retriver.invoke(query)
ret[0].page_content
```
